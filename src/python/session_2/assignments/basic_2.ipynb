```python
#Task One :
email = input("Enter your email address")
if "@" in email and "." in email.split("@")[-1]:
    print("Valid email address")
    username = email.split("@")[0]
    print("Username:", username)
    domain = email.split("@")[-1]
    print("Domain:", domain)
    if domain.endswith(".com"):
        print("This is a commercial email address")
    elif domain.endswith(".edu"):
        print("This is an educational email address")
    else:
        print("This is a different type of email address")
else:
    print("Invalid email address")

    
```

```python
#TAsk Two :
msg = "###!!@mocleW EPGTQ!!!6789"

core = msg[6:18]
print(core)

first_word = core[:6][::-1]
print(first_word)

second_word = core[7:12]
print(second_word)

letters = list(second_word)
letters = [ch for ch in letters if ch != "E"]
second_word = "".join(letters)
print(second_word)

print(first_word, second_word) 
```
